## Function which returns two groups in the input:

```python
#Write a regex match function 
#which takes each record and returns the list with two items
#item1: capture group 1
#item 2: capture group 2

def reg_func(string):
    charReg = re.search(r'\(?([\w,\&\-\s]*),\s(\d{4})\)?',string)
    return("The First group is {} and the second group is {}.".format(charReg.group(1),charReg.group(2)))
```

```python
reg_func(string)
```

```python
def match_author(record):
    author_match = re.match(auth_regex,record)
    authors = author_match.group(1)
    year = author_match.group(2)
    authors = re.split(',|\&',authors)
    #This splits the authors via ,|& into different names
    count = defaultdict(int)
    for ii in authors:
        count[ii]=count[ii]+1
    return (count)
```
