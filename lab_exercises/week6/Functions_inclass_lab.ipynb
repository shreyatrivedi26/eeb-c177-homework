```python
def fahr_to_cel(temp):
    return((temp-32)*(5/9))
```

```python
cel_75= fahr_to_cel(75)
print("75F in Celcius is: ",cel_75,"C")
```

```python
def celcius_to_kelvin(temp_c):
    return(temp_c+273)
```

```python
print("23.888C in Kelvin is: ",celcius_to_kelvin(23.888),"K")
```

```python
def fahr_to_kel(temp_f):
    temp_c = fahr_to_cel(temp_f)
    temp_k = celcius_to_kelvin(temp_c)
    return temp_k
```

```python
print(fahr_to_kel(75))
```

```python
def visualize(filename,delim=','): #by defining the default values, such as comma here, if we are not proving delimiter everytime, when not provided it will take comma by default
    #Goal: to create a function that will do the foll steps:
    #1) Read in the data files
    data = numpy.loadtxt(fname = filename, delimiter = delim) #filename and delimiter not defined but kept as inputs
    
    #2) create our three subplot holders
    fig = matplotlib.pyplot.figure(figsize=(10.0,3.0))
    axes1= fig.add_subplot(1,3,1)
    axes2= fig.add_subplot(1,3,2)
    axes3= fig.add_subplot(1,3,3)
    
    #3) plot each of them 
    axes1.set_ylabel('Average')
    axes1.plot(numpy.mean(data,axis=0))
    
    axes2.set_ylabel('Maximum')
    axes2.plot(numpy.max(data,axis=0))
    axes2.set_xlabel('No. of Days')
    
    axes3.set_ylabel('Minimum')
    axes3.plot(numpy.min(data,axis=0))
    
    #4) we want to show the plot
    fig.tight_layout
    matplotlib.pyplot.show()
    
```

```python
def offset_mean(data, target_mean_value):
    
    '''Return a new array contatining the original data with its mean offset to match the desired value, (0 by default)
    
       Examples
       --------
       >>> offset_mean([1,2,3])
       array([-1,0,1])
       '''
    
    return(data-numpy.mean(data)) + target_mean_value
```

## `while` Loops (Indefinite Loops) 

While loops run endlessly till the condition is met. It uses the advantage of booleans (True and False) datasets. 
 

```python
n = 5
while n > 0:
    n-=1
    print(n)
```

```python
n = 5
while n > 0:
    print("Before decrement: ",n)
    n-=1
    print("After decrement: ",n)
```

```python
a = ['foo','bar','baz']
while a:
    print(a.pop(-1)) #while loop ends as the elemenst are removed from the end. 
```

```python
n = 5 
while n>0:
    n-=1
    if n ==2:
        break
    print(n)
print('loop ended')
```

```python
n = 5 
while n>0:
    n-=1
    if n ==2:
        continue #2 will not print as the loop goes back and skips print 2
    print(n)
print('loop ended')
```

Python is the only language which allows the use of `else` with `while`.

```python
n = 5 
while n>0:
    n-=1
    print(n)
else:
    print('loop ended')
```
