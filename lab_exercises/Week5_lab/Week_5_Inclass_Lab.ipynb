## Memory allocation (Enumerate):

Shallow copy: A reassignment of the same variable which gets stored in the same memory location. 
    Eg. salsa = my_Salsa


Deep Copy: The duplication of the content but with a slight change which gets stored in a different memory location.
    Eg. my_salsa = list(salsa) 

```python

salsa = ["onions","cilantro","tomatoes","peppers"]
my_salsa = salsa 
print("salasa1: ", salsa)

#Shallow copy:

#print(enumerate(salsa))
#print(enumerate(my_salsa))

#Deep copy:

My_salsa = list(salsa)
print("salsa2:",My_salsa)
#print(enumerate(My_salsa))

#changing salsa:
salsa[3] = ['hot peppers']

#Changing the salsa will not change deep memory salsa but only shallow one
```

```python
odds=[3,5,7,9]
odds.reverse()
print(odds)
```

```python
#Reverse it back using for loop: [5,7,9]

odds_in_order = [] #arrays are dynamically defined irrespective of their size 
length_of_odds = len(odds)-1
for ii in range(0,len(odds)): #runs from 0 to the length of the array 
    odds_in_order.append(odds[length_of_odds-ii]) #appends the value of ii at the end of the array 
    
print("original reversed list:",odds_in_order)


#Second way:

odds_in_order = [] #arrays are dynamically defined irrespective of their size 
for ii in range(0,len(odds)): #runs from 0 to the length of the array 
    odds_in_order.append(odds[(len(odds)-ii)-1]) #appends the value of ii at the end of the array 
    
print("manual reversed list:",odds_in_order)

#Third way:

odds_in_order = [] #arrays are dynamically defined irrespective of their size 
for ii in range(0,len(odds)+1): #runs from 0 to the length of the array 
    odds_in_order.append(odds[len(odds)+1-ii]) #appends the value of ii at the end of the array 
    
print("flipped manual reversed list:",odds_in_order)
```

```python
#get all even numbers between 1 to 100: 
evens=[]
for ii in range (1,101):
    if(ii%2==0):
        evens.append(ii)

print ("Even numbers: ", evens)
```

```python
#get all odd numbers between 1 to 100: 
odds=[]
for ii in range (1,101):
    if(ii%2==1):
        odds.append(ii)

print ("Odd numbers: ", odds)

#or



odds=[]
for ii in range (1,101):
    if(ii%2!=0):
        odds.append(ii)

print ("\nOdd numbers: ", odds)
```
